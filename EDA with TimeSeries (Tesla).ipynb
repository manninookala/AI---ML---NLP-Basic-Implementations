```python
df_tesla['Open'].resample(rule = 'BA').mean().plot(kind = 'bar')
```

```python
df_tesla['Open'].resample(rule = 'M').mean().plot(kind = 'bar', figsize = (12,6))
```

# **Simple Moving Average (SMA) :**

```python
df_tesla['Open:10 days rolling'] = df_tesla['Open'].rolling(window=10, min_periods=1).mean()
df_tesla['Open:20 days rolling'] = df_tesla['Open'].rolling(window=20, min_periods=1).mean()
df_tesla['Open:30 days rolling'] = df_tesla['Open'].rolling(window=30, min_periods=1).mean()
df_tesla['Open:40 days rolling'] = df_tesla['Open'].rolling(window=40, min_periods=1).mean()
df_tesla['Open:50 days rolling'] = df_tesla['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df_tesla.head(15)
```

```python
df_tesla[['Open','Open:10 days rolling','Open:30 days rolling','Open:50 days rolling']].plot(xlim=['2022','2023'],figsize=(16,6))
```

# **Cummulative Moving Average (CMA) :**

```python
df_tesla['Open:CMA'] = df_tesla['Open'].expanding().mean()
```

```python
df_tesla[['Open:CMA','Open']].head(20)
```

```python
df_tesla[['Open','Open:CMA']].plot(xlim=['2017','2022'])
```

```python
df_tesla['Open'].expanding().mean().plot()
```

# **Exponential Moving Average :**

```python
df_tesla['Open:EMA0.1'] = df_tesla['Open'].ewm(alpha=0.1, adjust = False).mean()
df_tesla['Open:EMA0.3'] = df_tesla['Open'].ewm(alpha=0.3, adjust = False).mean()
df_tesla['Open:EMA0.7'] = df_tesla['Open'].ewm(alpha=0.7, adjust = False).mean()
```

```python
df_tesla[['Open','Open:EMA0.1','Open:EMA0.3','Open:EMA0.7']].plot(xlim=['2021','2022'], figsize=(14,6))
```

```python
df_tesla['Open:EWMA5'] = df_tesla['Open'].ewm(span=5).mean()
df_tesla['Open:EWMA15'] = df_tesla['Open'].ewm(span=15).mean()
df_tesla['Open:EWMA25'] = df_tesla['Open'].ewm(span=25).mean()
```

```python
df_tesla[['Open','Open:EWMA5','Open:EWMA15','Open:EWMA25']].plot(xlim=['2021','2022'], figsize=(14,6))
```
