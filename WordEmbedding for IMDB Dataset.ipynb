```python
from tensorflow.keras.preprocessing.text import one_hot
voc_size = 3000
```

```python
ohe_rep = [one_hot(word,voc_size)for word in sent]
ohe_rep[:3]
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length = 3000
```

```python
pad_doc = pad_sequences(ohe_rep, padding='post', maxlen=sent_length)
pad_doc[:3]
```

```python
dim = 250
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim, input_length = sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(pad_doc)
```

```python
vectors = model.predict(pad_doc)
```

```python
print(len(vectors))
```

```python
flattened_vectors = vectors.flatten()
```
