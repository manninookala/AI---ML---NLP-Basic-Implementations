```python
df_subset = df.loc['2017-05-11':'2022-05-10']
```

```python
df_subset['High'].max()
```

```python
df_subset['High'].idxmax()
```

```python
df_subset['Low'].idxmin()
```

```python
max_price_row = df_subset['High'].idxmax()

# Get the maximum price
max_price = df_subset.loc[max_price_row, 'High']

print(f"The highest price of Microsoft's stock between 2017 and 2022 was on {max_price_row}, with a price of {max_price}.")
```

```python
min_price_row = df_subset['Low'].idxmin()

# Get the maximum price
min_price = df_subset.loc[min_price_row, 'Low']

print(f"The highest price of Microsoft's stock between 2017 and 2022 was on {min_price_row}, with a price of {min_price}.")
```
